```python
t = dat3['age']
dat3['age'] = (t - t.min())/(t.max()-t.min())
```

```python
dat3
```

```python
t = dat3['fecha_alta']
```

```python
mint = t.min()
maxt = t.max()
result = []
for i in t:
    i = pd.Timestamp(i)
    n = i - pd.Timestamp(mint)
    m = pd.Timestamp(maxt) - pd.Timestamp(mint)
    result.append((n.days+0.0)/m.days)
```

```python
dat3['fecha_alta'] = np.array(result)
```
